```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
data ={
    "house_cat":[1,2,3,4,5],
    "house_price":[100,2000,3000,40005,50000]
}
res = pd.DataFrame(data)

model = IsolationForest(contamination=0.2,random_state=42)
outliers=model.fit_predict(res[["house_price"]])
res["Is_outlier"]=np.where(outliers==-1,True,False)
res
```
